# BAHANG
Bahang merupakan sebutan lain dari panas, baik yang tersimpan maupun yang dirasakan, yaitu energi yang dipindahkan dari suatu obyek ke obyek lain karena perbedaan suhu. Bahang yang dapat dirasakan mampu diindra dengan kulit kita sebagai rangsang panas dan dingin. Jika kita ingin lebih teliti, banyaknya bahang yang mampu dirasakan dapat diukur dengan termometer. Bahang tersembunyi tak dapat kita ukur, namun ia tampil ketika wujud benda berubah. 
Proses pembentukan awan merupakan salah satu peristiwa meteorologi yang melibatkan bahang. Bahang diambil oleh air dari radiasi matahari, lalu disimpan bilamana ia berubah menjadi uap. Proses pelepasan bahang terjadi ketika uap air dinaikkan oleh proses konveksi, lalu mendingin menjadi titik-titik air yang kita bisa kita amati sebagai awan. Awan pun akan terus tumbuh jika bahang yang dilepaskan uap air di sana cukup untuk melanjutkan proses konveksi.

## Suhu Udara
Dalam kehidupan sehari-hari, kita biasa menerangkan seberapa panas suatu benda dengan menyatakannya dalam satuan suhu. Ada banyak satuan suhu yang bisa kita pakai. Di sebagian besar belahan dunia, satuan Celsius umumnya dipakai untuk menggambarkan suhu. Satuan Celsius pertama kali dibuat oleh ahli astronomi dari Swedia, Anders Celsius tahun 1742, 2 tahun sebelum kematiannya. 0 derajat Celsius saat itu didefinisikan sebagai titik didih air dan 100 derajat Celsius sebagai titik beku air, semuanya dalam tekanan atmosfer standar (1 atm atau 101.325 KPascal) dan diukur menggunakan termometer air raksa. Di Prancis, Jean Pierre Christin pada tahun 1743 membuat skala yang sama dengan Celsius, namun dengan definisi 0 derajat dan 100 derajat yang dibalik. Karena alasan etika, definisi skala yang dibuat Anders Celsius baru dibalik oleh ilmuwan Swedia setelah kematiannya. Saat ini, definisi skala Celsius diikatkan ke skala Kelvin.

Di kawasan Amerika Serikat, skala Fahrenheit umumnya digunakan untuk menerangkan suhu. Skala Fahrenheit dibuat pada tahun 1724 oleh Daniel Gabriel Fahrenheit di Polandia, 18 tahun sebelum skala Celsius. Fahrenheit menyatakan 32 derajat Fahrenheit sebagai titik beku air, 0 derajat sebagai suhu campuran es dan garam, dan 96 derajat sebagai suhu tubuh manusia pada umumnya; semua benda tersebut diukur menggunakan termometer air raksa. Setelah Fahrenheit meninggal pada tahun 1736, para ilmuwan merevisi skala Fahrenheit dengan menambahkan definisi 212 derajat Fahrenheit sebagai titik didih air, sedemikian sehingga suhu tubuh manusia dinyatakan sebagai 98,6 derajat Fahrenheit hingga saat ini. 
Masih banyak lagi skala suhu yang bisa kita pakai, seperti skala Rankine, Rømer, Newton, Planck, dan lain-lain. Bahkan para koki dan ibu-ibu rumah tangga menggunakan skala suhu berdasarkan setelan oven. Kita pun juga bisa membuat skala suhu berdasarkan preferensi kita sendiri.

Banyaknya skala suhu yang tercipta mendorong perlunya suatu satuan suhu yang absolut, yang diciptakan berdasarkan hukum-hukum fisika dan perhitungan matematis dari peristiwa-peristiwa fisika terkontrol di laboratorium. Sebagai satuan suhu yang absolut, dia harus bersifat konsisten sehingga apabila satuan ini diaplikasikan pada beragam jenis termometer hasilnya harus selalu sama, begitu pula apabila hasil pengukuran suhu tadi dimasukkan ke berbagai persamaan dan operasi matematis.
Karena sulitnya membuat suatu skala suhu yang absolut, hal itu baru tercapai pada tahun 1848. Lord Kelvin, dalam makalahnya yang berjudul On an Absolute Thermometric Scale ‘Mengenai Skala 

Termometer yang Absolut’, mengutarakan bahwa skala termometer absolut dapat dibuat dari termometer udara. Kelvin saat itu mengutip hasil percobaan Regnault yang menggunakan termometer udara saat mengukur suhu (dalam skala Celsius), yang menunjukkan hasil pengukuran yang tidak berbeda jauh walaupun udara dalam termometer diubah komposisi dan tekanannya. Menggabungkan hasil percobaan Regnault dan teori Carnot mengenai “daya gerak dari bahang”, Kelvin mengasumsikan bahwa ‘jika udara dalam termometer udara menyusut hingga tiada’ sehingga ‘mencapai suhu paling dingin yang bisa dicapai’ maka suhu itu adalah -1⁄((0.0366 ℃) )=-273 ℃, 0.0336 1/˚C adalah koefisien muai udara (pada suhu 0 ˚C). Lord Kelvin menyatakan juga bahwa tidak ada termometer satu pun yang bisa mencapainya. Karena itulah, titik tersebut dinyatakan sebagai titik nol mutlak untuk skala Kelvin. 
Pada tahun 1969, didefinisikan bahwa satu Kelvin adalah 1/273,16 dari suhu air pada [titik tripel](https://www.calculator.org/properties/triple_point_temperature.html) (titik pada grafik suhu – tekanan yang menggambarkan parameter yang memungkinkan air berada pada wujud campuran antara beku, cair, dan gas). Suhu titik tripel air ditetapkan sebagai suhu 273,16 K atau 0,01˚C. Melalui percobaan, tekanan udara yang memungkinkan titik tripel air terjadi adalah 611,657 Pascal. Titik nol Kelvin terletak pada -273,15 ˚C. Kenaikan suhu pada skala Kelvin ditentukan sama seperti kenaikan suhu pada skala Celsius, sebagaimana yang telah ditegaskan Lord Kelvin pada makalahnya.
 
![](https://upload.wikimedia.org/wikipedia/commons/thumb/3/34/Phase-diag2.svg/500px-Phase-diag2.svg.png)

Gambar 34. Titik tripel digambarkan dalam grafik suhu – tekanan.

Skala Kelvin banyak digunakan untuk kepentingan ilmiah. Agar dapat dimengerti khalayak, skala Kelvin perlu dikonversi ke dalam skala Celsius dan Fahrenheit. Konversi dari skala Kelvin ke skala Celsius dapat terlaksana dengan perumusan berikut. 	
$T_C=T_K-273.15^o$ (4.1)

Sedemikian sehingga jika kita ingin mengonversi suhu yang dinyatakan dalam skala Celsius atau Kelvin ke dalam skala Fahrenheit, maka dapat menggunakan persamaan berikut.
$ T_F=9/5 T_C+32^o$ 	(4.2)
$ T_F=9/5 (T_K-273.15^o)+32$  (4.3)

Teori Kinetik Gas, yang konsep awalnya telah dibangun oleh Robert Boyle pada awal abad ke-17. Ia menyatakan bahwa suatu bejana berisi udara cenderung melawan jika ditekan, karena dua kemungkinan:
1. Udara terdiri atas partikel yang saling melawan, mirip seperti wol atau pegas.
2. Udara terdiri atas partikel yang bergerak memusar, yang saling tolak-menolak karena bertabrakan.

Teori Kinetik Gas ini terus berkembang seiring berjalannya waktu. Pada akhir abad ke-18, Boltzmann dan Maxwell mengajukan suatu perumusan yang menyatakan bahwa suhu udara (T, dalam Celsius atau Kelvin) bergantung pada energi kinetik geser rata-rata dari udara dalam suatu tempat. Energi kinetik geser adalah energi kinetik yang berkaitan dengan pergeseran molekul (bukan rotasi molekul). Perumusan energi kinetik geser rata-rata dari udara (K_trans ) ̅ (dalam kJ) adalah sebagai berikut.
$ \bar{K_{trans}} =  3/2 k T $  (4.4)

Huruf k pada rumus di atas merupakan tetapan Boltzmann yaitu $1.381 \times 10^{-23} $ Joule / Kelvin.

Walaupun dalam meteorologi, suhu udara lebih ditentukan oleh peristiwa-peristiwa termodinamika berskala besar – yang perumusannya merupakan pengembangan dan modifikasi dari teori kinetik gas beserta hasil eksperimen – namun persamaan (4.4) ini cukup bermanfaat dalam memahami peran bahang (kalor) sebagai bentuk energi dalam meningkatkan suhu dan perubahan wujud benda. Contohnya, ketika kita membicarakan tentang penguapan air laut, yang molekul-molekul airnya dapat melarikan diri ke udara berkat penambahan energi kinetik molekul dari panas matahari.

## Bahang Jenis dan Bahang Laten

Dalam kehidupan sehari-hari, kita mengenal adanya perpindahan panas dari sumbernya menuju tempat yang lebih dingin dan mustahil berlaku sebaliknya (Hukum Termodinamika Kedua). Adanya input bahang di tempat yang dingin secara otomatis meningkatkan energi internal benda-benda di tempat tersebut hingga tingkat tertentu sesuai jumlah energi yang masuk. Salah satu tanda adanya peningkatan energi internal dapat dimaknai dari peningkatan suhu atau perubahan wujud benda.
Bahang biasa dilambangkan dengan $Q$. Hubungannya dengan perubahan suhu $\Delta T$ (dalam Kelvin atau Celsius), massa benda $m$ (dalam kilogram), dan bahang (kalor) jenis benda $c$ (dalam kJ/kg∙K) dapat digambarkan oleh persamaan berikut.

$ Q=mc\Delta T$  (4.5)

**Bahang jenis** sendiri dapat dimaknai sebagai jumlah perubahan energi dalam yang dibutuhkan untuk menaikkan suhu satu kilogram benda tersebut sebanyak satu Kelvin atau satu Celsius, tanpa adanya perubahan wujud benda. Dalam persamaan (4.6), $C$ adalah kapasitas bahang yang dibutuhkan benda (kJ/K atau kJ/C) dan $m$ adalah massa benda dalam kilogram.
$ c=C/m $ (4.6)

Dalam proses perubahan wujud benda akibat pemanasan, terjadi hal yang menarik; bahang masuk namun suhu benda tidak bertambah. Hal itu terjadi karena hampir tidak ada bahang yang digunakan untuk menambah energi kinetik molekul benda saat proses perubahan wujud. Bahang yang masuk dipergunakan untuk menambah energi potensial setiap molekul, sehingga gaya tarik antar molekul menjadi rusak dan lebih lemah dari sebelumnya. Proses pelemahan gaya tarik antar molekul ini terejawantahkan dalam perubahan wujud molekul; dan bahang yang terlibat dalam proses tersebut dinamai bahang laten (tersembunyi).
Total bahang $Q_f$ yang dilepaskan saat benda membeku sama dengan bahang yang diperlukan untuk mencairkan atau menyublimkan benda tersebut. Bedanya terletak pada usaha yang dilakukan – usaha yang diperlukan untuk membekukan benda selalu melebihi usaha yang diperlukan untuk mencairkan atau menyublimkan benda. Begitu pula ketika mengubah wujud suatu benda menjadi uap, total bahang $Q_V$ yang diperlukan sama dengan ketika mengubah benda tadi kembali menjadi padat/cair. Usaha untuk menguapkan tentu saja lebih kecil dari usaha untuk mencairkan atau memadatkan kembali uap. Perumusan bahang $Q_f$ dan $Q_V$ adalah seperti berikut. 
$ Q_f=mL_f$ (4.7)
$ Q_V=mL_V$ (4.8)

$L_f$ merupakan bahang laten untuk proses perubahan wujud dari padat ke cair/gas dan sebaliknya, sementara $L_V$ merupakan bahang laten untuk proses perubahan wujud dari cair/padat ke gas. $L_f$ dan $L_V$ dinyatakan dalam satuan kJ/kg. Seperti biasa, m merupakan massa benda dalam kilogram.
Dalam meteorologi yang akan kita pelajari, konsep bahang jenis dan bahang laten akan dipakai terutama saat kita membahas mengenai kelembaban, presipitasi, dan proses-proses cuaca yang terkait dengan mereka.
## Perpindahan Bahang
### Konduksi
Perpindahan bahang dengan metode konduksi terjadi ketika satu bagian massa dipanaskan, maka bahang akan merambat ke seluruh bagian massa setelah beberapa waktu. Hal ini terjadi karena ketika ujung massa dipanaskan, energi kinetik molekul massa di bagian itu akan meningkat sehingga getarannya semakin kuat. Getaran molekul dari ujung massa mengganggu molekul-molekul di bagian massa yang lain, sehingga suhu massa akan naik. Namun, naiknya suhu massa ini tidak merata; tentu saja kenaikan suhu lebih cepat di bagian massa yang dekat sumber bahang, karena adanya peredaman getaran molekul oleh molekul lainnya. Besarnya peredaman ini beragam pada setiap benda, sehingga muncullah istilah konduktor baik dan konduktor buruk.
Logam, merupakan konduktor yang baik, karena selain oleh getaran molekul, perambatan panas pada logam juga dibantu oleh elektron bebas. Udara merupakan contoh konduktor yang buruk. Ketika matahari menyinari bumi, hanya selapis tipis udara paling bawah yang terpanasi langsung oleh tanah.
Selain dipengaruhi oleh konstanta peredaman, perambatan bahang melalui konduksi juga dipengaruhi oleh perbedaan suhu antara ujung panas dan ujung dingin massa. Semakin besar perbedaan suhu antara dua ujung tersebut, maka perambatan bahang semakin cepat. Perumusan laju perambatan bahang telah dihasilkan sejak abad ke 18 oleh Fourier. Perumusan itu kemudian disempurnakan seiring waktu, dengan bantuan eksperimen fisika terkontrol.
Laju perambatan bahang dapat diwakilkan oleh sebuah vektor q ⃗. Karena ketika bahang merambat ia tentu punya besaran (seberapa cepat ia merambat) dan arah (kemanakah ia merambat).
$\vec{q}=k∙A∙\vec{∇}T$	(4.9)

$\vec{q}$  merupakan laju perambatan bahang dalam Joule per detik, $k$ merupakan konduktivitas termal, bisa dianggap sebagai wakilan dari “konstanta peredaman getaran”, A merupakan luasan medium yang dilewati bahang, dan $\vec{∇}T$ merupakan vektor yang menunjukkan arah dan besar perubahan suhu dalam medium konduksi (gradien suhu).

### Radiasi
Perpindahan bahang dengan radiasi banyak bermanfaat dalam kehidupan sehari-hari kita. Pada siang hari, gelombang elektromagnetik yang dipancarkan proses-proses termonuklir matahari memanaskan Bumi setelah melewati jagat raya yang hampir vakum. Pada malam harinya, kita merasa hangat di lapangan berkat radiasi api unggun yang kita nyalakan.
Awalnya mungkin kita cukup bingung, mengapa cahaya yang massanya mendekati nol mampu membawa energi yang sangat besar? Secara mudah, kita bisa mengacu ke makalah-makalah Einstein tahun 1905 yang menyatakan bahwa cahaya memiliki energi yang bergantung pada frekuensinya, dan energi juga merupakan bentuk lain dari massa. Maka dari itu, tidak heran bila cahaya dapat membawa energi dalam jumlah besar walau massa terukurnya mendekati nol, karena energi dalam cahaya itulah “massa sebenarnya”.
Perumusan Einstein yang menyatakan bahwa energi cahaya bergantung pada frekuensinya adalah
$ E=h∙f $ (4.10)

Perumusan (4.10) dicapai setelah ia mengamati hasil eksperimen efek fotoelektrik. Kemudian, rumus di bawah merupakan rumus yang sangat terkenal dari Einstein (semua orang terpelajar pasti kenal).
$ E=m∙c^2$	(4.11)

Rumus (4.11) adalah rumus yang menerangkan hubungan antara massa dan energi. Selain itu, rumus (4.11) juga sangat terkenal karena membuat beberapa negara termotivasi untuk membuat bom atom. Dalam rumus-rumus (4.10) dan (4.11) di atas, h merupakan tetapan Planck yang memiliki nilai yang sangat kecil, sebesar $ 6.626 \times 10^{-34}$ J∙s, frekuensi f merupakan frekuensi gelombang elektromagnetik, dalam Hertz atau 1/detik. Huruf m pada rumus (K.) merupakan massa (dalam kilogram), dan c merupakan kelajuan cahaya dalam ruang vakum (dalam meter per detik).
Sebagaimana matahari dan api unggun, semua benda yang suhunya di atas 0 K pasti memancarkan radiasi elektromagnetik. Selain memancarkan gelombang elektromagnetik, benda juga menyerap energi gelombang elektromagnetik yang datang padanya. Contohnya, tubuh kita akan memanas jika disinari matahari (menyerap energi sinar), dan akan memancarkan panas bilamana kita berada di ruangan gelap (memancarkan gelombang elektromagnetik). Jumlah energi gelombang elektromagnetik yang dapat dipancarkan dan diserap suatu benda bergantung pada suhu, bahan, warna dan kekilapan benda.
Josef Stefan dan Ludwig Boltzmann memiliki kontribusi dalam menemukan hubungan antara suhu, warna, bahan dan kekilapan benda. Hubungan tersebut dapat dirumuskan sebagai berikut.
$ P=eσ∙A∙T^4$	(4.12)

Pada rumus (4.12), $P$ merupakan bahang yang diserap atau dipancarkan benda setiap satu detik (satuan Joule per detik), $e$ merupakan emisivitas benda (pemancar maupun penyerap), bergantung pada warna dan kekilapan benda, $A$ merupakan luasan benda yang memancarkan atau menyerap bahang (dalam meter persegi), dan $T$ merupakan suhu benda dalam Kelvin.

### Konveksi

Proses konveksi merupakan proses pemindahan bahang ke atas dengan mengikutkan media yang menampungnya. Proses perpindahan bahang ini akan banyak di bahas dalam buku ini, karena menyangkut proses pembentukan awan dan hujan. Selain itu, proses konveksi juga dapat dilihat ketika kita merebus air dan kacang; gerakan arus konveksi terlihat dari naik turunnya butiran kacang dalam air rebusan.
![](Gambar/ilustrasi_fluida.jpg) 

Gambar 35. Ilustrasi kondisi fluida yang memungkinkan terjadinya konveksi. Massa jenis fluida ρ’ kurang dari massa jenis fluida ρ, sehingga tekanan hidrostatik p- kurang dari tekanan hidrostatik p+. (sumber: MIT Program in Atmospheres, Oceans, and Climate).

Konveksi terjadi apabila suatu “tiang” fluida ρ dipanaskan bagian dasarnya (oleh api, tanah, atau yang lainnya). Akibatnya, timbul gelembung-gelembung fluida ρ‘ di dasar wadah akibat pemanasan yang tidak merata. 
$p_+=P_0+\int_{B_1}^{A_1} g∙ρ(z)dz <p_-=P_0+\int_B^A g∙ρ(z)dz $ 	(4.13)

Gelembung-gelembung fluida ρ‘ merupakan fluida yang komposisi kimianya sama dengan fluida ρ namun dengan massa jenis yang lebih rendah dan suhu yang lebih tinggi. Massa jenis ρ’ kurang dari ρ sehingga tekanan hidrostatik di bawah gelembung fluida ρ’ (p-) kurang dari tekanan hidrostatik di bawah fluida ρ (p+), maka fluida pada p+ akan bergerak menuju p-, sebagai implikasi dari Hukum Newton Kedua dan Hukum Bernoulli. Fluida ρ’ kemudian naik seiring masuknya fluida ρ ke bawahnya. Percepatan naiknya fluida ρ’, b, dapat dituliskan dalam persamaan berikut.
	b=-g(ρ^'-ρ)/ρ'	(4.14)
Ketika gelembung fluida ρ’ naik, ia juga akan membawa naik bahang yang dikandungnya. Sesuai Hukum Termodinamika Kedua, bahang ini akan dilepaskan ke fluida ρ dan lingkungan yang melingkupinya, sehingga gelembung fluida ρ’ akan mendingin selama perjalanan naik.
Dalam persamaan (M.) dan (N.), g merupakan percepatan gravitasi di suatu tempat, P0 merupakan tekanan udara atau beban lain yang menindih serta melingkupi fluida, ρ(z) menggambarkan perbedaan jenis fluida – dilihat dari massa jenisnya – sebagai fungsi kedalaman (z), ρ^' dan ρ merupakan nama dari dua jenis fluida yang massa jenisnya berbeda.
### Adveksi
Proses perpindahan bahang dengan perantaraan medium dalam arah horizontal disebut adveksi. Hal ini bisa kita amati ketika mempelajari proses pembentukan badai petir di dataran rendah dekat laut. Pada siang hari, angin laut bertiup dari perairan menuju daratan, membawa serta uap air dan bahang yang ada padanya. Ketika uap air dari laut ini diangkat oleh proses konveksi di daratan, mereka akan berkondensasi dan melepaskan bahang laten mereka menjadi bahang jenis. Karena itulah, beberapa waktu menjelang hujan kita akan merasa gerah.
Proses adveksi juga terjadi di Lautan Atlantik. Arus Teluk di Pantai Timur Amerika membawa air laut bersuhu hangat dari kawasan tropis menuju Eropa Utara. Bahang dari Arus Teluk dilepaskan ke udara selama perjalanannya, sehingga cuaca di Eropa Utara tidak terlampau dingin walaupun dekat dengan kutub. Sebaliknya, arus laut yang memutari Antartika memiliki suhu yang dingin, sehingga melindungi kutub selatan dari pencairan es besar-besaran. Hanya sedikit adveksi air laut tropik ke kutub selatan.
Secara matematis, adveksi dari bahang Q dapat dituliskan sebagai v ⃗∙∇ ⃗Q, dengan v ⃗ menunjukkan kecepatan medium pembawa bahang, dan ∇ ⃗Q merupakan vektor yang menunjukkan arah dan besar perubahan bahang dalam medium.
## Pemanasan Udara sebagai Proses Cuaca
Hukum pertama termodinamika menyatakan bahwa jumlah perubahan energi yang ada di dalam suatu sistem sama dengan total bahang yang berpindah ke dalamnya ditambah kerja yang dilakukan kepadanya. Dalam konteks meteorologi, sistem ini merupakan kolom udara yang terletak di atas suatu wilayah, dengan perpindahan bahang berasal dari tanah, angin, atau konveksi udara. Kerja dapat diwakili oleh pemanasan yang dilakukan sinar matahari ke kolom udara di sana. Pernyataan-pernyataan di atas dapat dituliskan ke dalam bahasa matematis seperti berikut.
	q=c_v (∂T/∂t+v ⃗∙∇ ⃗T)+p(∂α/∂t+v ⃗∙∇ ⃗α)	(4.15)

Dalam persamaan di atas, q adalah laju pemanasan udara, c_v merupakan bahang jenis untuk proses-proses termodinamika pada volume tetap, ∂T/∂t merupakan perubahan suhu udara setiap perubahan waktu, v ⃗∙∇ ⃗T merupakan lambang matematis untuk adveksi udara bersuhu T dengan kecepatan perpindahan v ⃗, p merupakan tekanan udara, α merupakan volume jenis udara, yaitu angka yang menunjukkan banyaknya volume dalam 1 kg udara.
